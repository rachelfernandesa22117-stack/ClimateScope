```python
#Removed Redundant Units
existing_columns = [col for col in [
    'temperature_fahrenheit',
    'wind_mph',
    'pressure_in',
    'precip_in',
    'feels_like_fahrenheit',
    'visibility_miles',
    "carbon_monoxide",
    "ozone",
    "nitrogen_dioxide",
    "sulphur_dioxide",
    'gust_mph'
]if col in df.columns]
```

```python
if existing_columns:
    df.drop(columns=existing_columns, inplace=True)
    print(f"Dropped columns: {existing_columns}")
else:
    print("None of the specified columns exist in the DataFrame")
```

```python
#Converted Date Columns
df['updated_datetime'] = pd.to_datetime(df['updated_datetime'])

# Extracted useful features
df['year'] = df['updated_datetime'].dt.year
df['month'] = df['updated_datetime'].dt.month
df['day'] = df['updated_datetime'].dt.day
df['hour'] = df['updated_datetime'].dt.hour
```

```python
#Average UV per Country
country_uv = df.groupby('country')['uv_index'].mean().reset_index()
```

```python
def country_uv_category(x):
    if x <= 2:
        return "Low UV Country"
    elif x <= 5:
        return "Moderate UV Country"
    elif x <= 7:
        return "High UV Country"
    else:
        return "Very High UV Country"
```

```python
country_uv['uv_risk_level'] = country_uv['uv_index'].apply(country_uv_category)
print(country_uv)
```

```python
country_name = "India"

avg_uv = df[df['country'] == country_name]['uv_index'].mean()

print(f"{country_name} has an average UV index of {avg_uv:.2f}, classified as {uv_category(avg_uv)} risk.")
```

```python
#Normalized Temperature
df['temperature_celsius_norm'] = (
    df['temperature_celsius'] - df['temperature_celsius'].min()
) / (
    df['temperature_celsius'].max() - df['temperature_celsius'].min()
)
```

```python
#Normalized Humidity
df['humidity_norm'] = (
    df['humidity'] - df['humidity'].min()
) / (
    df['humidity'].max() - df['humidity'].min()
)
```

```python
#Normalized UV Index
df['uv_index_norm'] = (
    df['uv_index'] - df['uv_index'].min()
) / (
    df['uv_index'].max() - df['uv_index'].min()
)
```

```python
print(df[['temperature_celsius', 'temperature_celsius_norm']])
```

```python
cols_to_normalize = [
    'temperature_celsius',
    'humidity',
    'precip_mm',
    'wind_kph',
    'pressure_mb',
    'visibility_km',
    'uv_index',
    'gust_kph',
]
```

```python
for col in cols_to_normalize:
    df[col + '_norm'] = (
        (df[col] - df[col].min()) /
        (df[col].max() - df[col].min())
    )
```

```python
#Monthly aggregation was performed using groupby on country, year and month
monthly_df = df.groupby(
    ['country', 'year', 'month']
).agg({
    'temperature_celsius':'mean',
    'humidity':'mean',
    'precip_mm':'sum',
    'wind_kph':'mean'
}).reset_index()
print(monthly_df.head())
```
